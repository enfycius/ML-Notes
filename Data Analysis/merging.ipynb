# Merging/Joining

-> 관계형 데이터베이스(Relational Databases) (SQL)와 매우 유사한 In-memory join operations을 Pandas가 제공해줌.

````python
import pandas as pd
````

````python
left = pd.DataFrame({
    "id": [1, 2, 3, 4, 5],
    "Name": ["Alex", "Amy", "Allen",
             "Alice", "Ayoung"],
    "subject_id": ["sub1", "sub2",
                   "sub4", "sub6", 
                   "sub5"]
})
````

````python
left
````

````python
right = pd.DataFrame({
    "id": [1, 2, 3, 4, 5],
    "Name": ["Billy", "Brian",
             "Bran", "Bryce",
             "Betty"],
    "subject_id": ["sub2", "sub4",
                   "sub3", "sub6",
                   "sub5"]
})
````

````python
right
````

```python
pd.merge(x, y)
```

````python
pd.merge(left, right, on="id")
````

## 길이가 서로 다를 때

````python
left = pd.DataFrame({
    "id": [1, 2, 3, 4, 5, 6],
    "Name": ["Alex", "Amy", "Allen",
             "Alice", "Ayoung", "AA"],
    "subject_id": ["sub1", "sub2",
                   "sub4", "sub6", 
                   "sub5", "sub9"]
})
````

````python
left
````

````python
right = pd.DataFrame({
    "id": [1, 2, 3, 4, 5],
    "Name": ["Billy", "Brian",
             "Bran", "Bryce",
             "Betty"],
    "subject_id": ["sub2", "sub4",
                   "sub3", "sub6",
                   "sub5"]
})
````

````python
right
````

```python
pd.merge(x, y)
```

````python
pd.merge(left, right, on="id")
````

> merge를 할 때, on 키워드에 들어가는 컬럼명을 기준으로 left와 right에 공통적으로 존재하는 여기서는 id값들 (1~5까지)에 대해서만 Merge가 된다.

````python
pd.merge(left, right, on="subject_id")
````

````python
print(left)
print(right)
````

````python
pd.merge(left, right, on=["id", "subject_id"])
````

> on 키워드에 지정한 공통된 컬럼들에 대해서 공통되는 값들이 서로 Merge가 되는 원리

# Joining

## Left Join

````python
pd.merge(left, right, on="subject_id", how="left")
````

````python
right = pd.DataFrame({
    "id": [1, 2, 3, 4, 5],
    "Name": ["Billy", "Brian",
             "Bran", "Bryce",
             "Betty"],
    "subject_id": ["aa1", "aa2",
                   "aa3", "aa4",
                   "aa5"]
})
````

````python
pd.merge(left, right, on="subject_id", how="right")
````

# Grouping

````python
import numpy as np
import pandas as pd
````

````python
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)
````

````python
df
````

````python
df.groupby("A").sum()
````

````python
df.groupby(["A", "B"]).sum()
````

````python
df.groupby(["B", "A"]).sum()
````
