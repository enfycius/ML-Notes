```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.randn(1000).shape
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))
```

```python
ts
```

```python
ts = ts.cumsum()
```

```python
ts
```

```python
ts.plot()
```

```python
df = pd.DataFrame(
    np.random.randn(1000, 4), index=ts.index, columns=["A", "B", "C", "D"]
)
```

```python
df
```

```python
df = df.cumsum()
```

```python
plt.figure()

df.plot()

plt.legend(loc="best")
```
