# Covariance

: 두 변수 간의 선형 관계인지를 파악하기 위한 용도로 사용됨.

: 두 변수가 선형의 관계이면, 공분산은 양수이거나 음수이다. 즉 0이 아니다.

```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series(np.random.randn(10))
t = pd.Series(np.random.randn(10))
```

```python
s.cov(t)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(s, t)
plt.ylim([-10, 10])
plt.xlim([-10, 10])
plt.show()
```
